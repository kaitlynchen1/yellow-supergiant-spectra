```python
header = r"""\begin{deluxetable*}{clccccccc}
\tabletypesize{\scriptsize}
\tablecaption{Common names, coordinates, host galaxies, and \Gaia measurements of 6,484 putative massive stars. $r_{est}$ from \citet{bailerjones18} is given for Galactic stars. Listed values of $G$ and $G_{BP}-G_{RP}$ are uncorrected for extinction.\label{tab:sample}}
\tablehead{\colhead{Common Name} & \colhead{R.A. [deg]} & \colhead{Dec [deg]} & \colhead{Host Galaxy} & \colhead{$r_{est}$ [kpc]} & \colhead{$G$ [mag]} & \colhead{$A_G$ [mag]} & \colhead{$G_{BP}-G_{RP}$ [mag]}} 
\startdata
"""
j=0
for i,row in df.iterrows():
    if j == 39:
        break
    else:
        j+=1
    common_name = row['CommonName']
    # Teff
    teff_mc = row['teff_mc']
    teff_upper = row['teff_upper']
    teff_lower = row['teff_lower']
    teff_positive_error = teff_upper - teff_mc
    teff_negative_error = teff_mc - teff_lower
    teff_formatted = f"${teff_mc:.0f}\substack{{+{teff_positive_error:.3f} // -{teff_negative_error:.3f}}}$"

    # Logg
    logg_mc = row['logg_mc']
    logg_upper = row['logg_upper']
    logg_lower = row['logg_lower']
    logg_positive_error = logg_upper - logg_mc
    logg_negative_error = logg_mc - logg_lower
    logg_formatted = f"${logg_mc:.3f}\substack{{+{logg_positive_error:.3f} // -{logg_negative_error:.3f}}}$"

    # Radius
    radius_mc = row['radius_mc']
    radius_upper = row['radius_upper']
    radius_lower = row['radius_lower']
    radius_positive_error = radius_upper - radius_mc
    radius_negative_error = radius_mc - radius_lower
    radius_formatted = f"${radius_mc:.3f}\substack{{+{radius_positive_error:.3f} // -{radius_negative_error:.3f}}}$"
    
    # lum
    lum_mc = row['lum_mc']
    lum_upper = row['lum_upper']
    lum_lower = row['lum_lower']
    lum_positive_error = lum_upper - lum_mc
    lum_negative_error = lum_mc - lum_lower
    lum_formatted = f"${lum_mc:.3f}\substack{{+{lum_positive_error:.3f} // -{lum_negative_error:.3f}}}$"

    # ifmg
    ifmg_mc = row['teff_mc']
    ifmg_upper = row['teff_upper']
    ifmg_lower = row['teff_lower']
    ifmg_positive_error = teff_upper - teff_mc
    ifmg_negative_error = teff_mc - teff_lower
    ifmg_formatted = f"${teff_mc:.3f}\substack{{+{teff_positive_error:.3f} // -{teff_negative_error:.3f}}}$"
    
    index=row['index']
    teff_pm=row['teff_pm']
    lum_pm=row['lum_pm']
    halpha_emission=row['halpha_emission']
    fyps=row['FYPS']
    binary=row['binary']
    
    
    entry = list(map(str,[index,common_name,ra,dec,gal,rest,g,ag,bmr]))
    
    sep = ' & '
    
    header += sep.join(entry)
    header += ' \\\\ \n'
header.rstrip()
header.rstrip('\\')
header.rstrip()
header += r"""\enddata
\tablecomments{The first ten rows are shown to illustrate the content of the table. A complete machine-readable version will be made available online via Vizier.}
\end{deluxetable*}
â€‹
"""
with open("sample_table.txt", "w") as text_file:
        text_file.write(header)
```
