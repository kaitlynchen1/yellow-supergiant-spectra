```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/kchen/Downloads/ysg_new.csv')
```

```python

for col in df.columns:
    print(col)
#del df['Unnamed: 0']
df.tail()
```

```python
header = r"""\begin{deluxetable*}{clcccccccccc}
\tabletypesize{\scriptsize}
\tablecaption{Index, Common Name, FYPS, binary status (0 for not binary, 1 for binary, 2 for maybe binary), H$\alpha$ in emission, our calculated log effective temperature, log photometric effective temperature, surface gravity, radius, our calculated luminosity, photometric luminosity, inverse flux mean gravity}
\tablehead{\colhead{index} & \colhead{CommonName} & \colhead{FYPS} & \colhead{Binary} & \colhead{H$\alpha$ in emission} & \colhead{log(Teff)} & \colhead{log(Teff PM)} & \colhead{log(g)} & \colhead{radius} & \colhead{lum} & \colhead{lum PM} & \colhead{ifwg}} 
\startdata
"""
j=0
for i,row in df.iterrows():
    if j == 40:
        break
    else:
        j+=1
    common_name = row['CommonName']
    # Teff
    teff_mc = row['teff_mc']
    teff_upper = row['teff_upper']
    teff_lower = row['teff_lower']
    teff_positive_error = 10*(teff_upper - teff_mc)
    teff_negative_error = 10*(teff_mc - teff_lower)
    teff_formatted = f"${teff_mc:.3f}\substack{{+{teff_positive_error:.3f} \\\\ -{teff_negative_error:.3f}}}$"

    # Logg
    logg_mc = row['logg_mc']
    logg_upper = row['logg_upper']
    logg_lower = row['logg_lower']
    logg_positive_error = 1*(logg_upper - logg_mc)
    logg_negative_error = 1*(logg_mc - logg_lower)
    logg_formatted = f"${logg_mc:.2f}\substack{{+{logg_positive_error:.2f} \\\\ -{logg_negative_error:.2f}}}$"

    # Radius
    radius_mc = row['radius_mc']
    radius_upper = row['radius_upper']
    radius_lower = row['radius_lower']
    radius_positive_error = 1*(radius_upper - radius_mc)
    radius_negative_error = 1*(radius_mc - radius_lower)
    radius_formatted = f"${radius_mc:.0f}\substack{{+{radius_positive_error:.0f} \\\\ -{radius_negative_error:.0f}}}$"
    
    # lum
    lum_mc = row['lum_mc']
    lum_upper = row['lum_upper']
    lum_lower = row['lum_lower']
    lum_positive_error = 10*(lum_upper - lum_mc)
    lum_negative_error = 10*(lum_mc - lum_lower)
    lum_formatted = f"${lum_mc:.3f}\substack{{+{lum_positive_error:.3f} \\\\ -{lum_negative_error:.3f}}}$"

    # ifwg
    ifwg_mc = row['ifwg_mc']
    ifwg_upper = row['ifwg_upper']
    ifwg_lower = row['ifwg_lower']
    ifwg_positive_error = 1*(ifwg_upper - ifwg_mc)
    ifwg_negative_error = 1*(ifwg_mc - ifwg_lower)
    ifwg_formatted = f"${ifwg_mc:.3f}\substack{{+{ifwg_positive_error:.3f} \\\\ -{ifwg_negative_error:.3f}}}$"
    
    index=row['index']
    teff_pm=f"{row['teff_pm']:.3f}"
    lum_pm=f"{row['lum_pm']:.3f}"
    halpha_emission=row['halpha_emission']
    fyps=row['FYPS']
    binary=row['binary']
    
    entry = list(map(str,[index,common_name,fyps,binary,halpha_emission,teff_formatted,teff_pm,logg_formatted,radius_formatted,lum_formatted,lum_pm,ifwg_formatted]))
    
    sep = ' & '
    
    header += sep.join(entry)
    header += ' \\\\ \n'
header.rstrip()
header.rstrip('\\')
header.rstrip()
header += r"""\enddata
\tablecomments{We obtained these results for the 39 spectra.}
\end{deluxetable*}
â€‹
"""
with open("sample_table.txt", "w") as text_file:
        text_file.write(header)
```
